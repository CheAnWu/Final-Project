```python
import pandas as pd
import numpy as np
```

```python
standpoint = pd.read_csv('BlackJackSimulation21_SplitFalse.csv')
```

```python
standpoint.head(20)
```

```python
Split = standpoint[standpoint['Player Hand Split'] > 0.0]
```

```python
Split.head(20)
```

```python
Split_out = Split[['Dealer Show','Player Hand','Game Result']]
```

```python
Split_out.head(20)
```

```python
Split_Result = pd.DataFrame(Split_out.groupby([Split_out['Player Hand'],Split_out['Dealer Show']]).sum(), columns=['Game Result']).reset_index()
Split_Count = pd.DataFrame(Split_out.groupby([Split_out['Player Hand'],Split_out['Dealer Show']]).size(), columns=['Count']).reset_index()
df_output_T = pd.merge(Split_Result, Split_Count, how='inner',on=['Player Hand', 'Dealer Show']).sort_values(['Player Hand', 'Dealer Show'], ascending=[True, True])
df_output_T['Expectation'] = df_output_T['Game Result']/df_output_T['Count']
```

```python
df_output_T
```

```python
df_output_T.to_csv('SplitFalse21.csv')
```
