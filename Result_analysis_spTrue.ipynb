```python
import pandas as pd
import numpy as np
```

```python
standpoint = pd.read_csv('BlackJackSimulation22.csv')
```

```python
standpoint.head(20)
```

```python
SplitTrue = standpoint[standpoint['split flag']==True]
SplitFalse = standpoint[standpoint['split flag']==False]
```

```python
TRUE = SplitTrue[['Player Hand','Dealer Show','Game Result']]
```

```python
FALSE = SplitFalse[['Player Hand','Dealer Show','Game Result']]
```

```python
TRUE_Result = pd.DataFrame(TRUE.groupby([TRUE['Player Hand'],TRUE['Dealer Show']]).sum(), columns=['Game Result']).reset_index()
TRUE_Count = pd.DataFrame(TRUE.groupby([TRUE['Player Hand'],TRUE['Dealer Show']]).size(), columns=['Count']).reset_index()
df_output_T = pd.merge(TRUE_Result, TRUE_Count, how='inner',on=['Player Hand', 'Dealer Show']).sort_values(['Player Hand', 'Dealer Show'], ascending=[True, True])
df_output_T['Expectation'] = df_output_T['Game Result']/df_output_T['Count']
```

```python
FALSE_Result = pd.DataFrame(FALSE.groupby([FALSE['Player Hand'],FALSE['Dealer Show']]).sum(), columns=['Game Result']).reset_index()
FALSE_Count = pd.DataFrame(FALSE.groupby([FALSE['Player Hand'],FALSE['Dealer Show']]).size(), columns=['Count']).reset_index()
df_output_F = pd.merge(FALSE_Result, FALSE_Count, how='inner',on=['Player Hand', 'Dealer Show']).sort_values(['Player Hand', 'Dealer Show'], ascending=[True, True])
df_output_F['Expectation'] = df_output_F['Game Result']/df_output_F['Count']
```

```python
df_output_T.head(20)
```

```python
df_output_T.to_csv('output_True_22.csv')
```
